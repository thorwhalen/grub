```python
@code_to_dag
def vectorized_search():
    query_vector = apply(embedder, query)
    result_keys = get_most_similar(query_vector, num_results, corpus_vectors, search_index)
    results = get_results(corpus, result_keys)

@code_to_dag
def prepare_corpus():
    corpus_vectors = apply(embedder, corpus)
    search_index = build_index(corpus_vectors)

@code_to_dag
def learn_embedding():
    embedder = apply(learner, learn_corpus)


search = vectorized_search + prepare_corpus + learn_embedding
search.dot_digraph()
```
